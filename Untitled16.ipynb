# Project Manifest: AI-Driven Roofing Measurement & Sales Platform

## 1. Development Protocol
**Framework**: `my_task = weld, paint, polish`
1.  **Weld**: Build the functional core (Data ingestion, Math, Basic Classes).
2.  **Paint**: Structure logic, API connections, and inter-component routing.

*   **Ubiquitous AI Chat**:

    *   Function: Heat scores, pipeline categorization, auto-quoting.

    *   Trigger: Pipeline states.
    *   Output: Drip campaigns, social media project summaries.

```python
class SalesPipeline:
    """
    The Intelligent Sales Pipeline (CRM).
    Role: Commercial Logic & Automation.
    Responsibility: Heat scoring, pipeline categorization, and auto-quoting.
    """
    def __init__(self, base_price_per_sq=450.00):
        self.base_price_per_sq = base_price_per_sq # $450 per square installed

    def calculate_heat_score(self, project):
        """
        Calculates a 'Heat Score' (0-100) to gauge lead value/urgency.
        Logic: Bigger roofs + reasonable complexity = Higher Score.
        """
        if not project.brain_output:
            return 0

        estimates = project.brain_output['estimates']
        squares = estimates['total_squares_recommended']

        # Scoring Logic
        # 1. Size Score: Ideal range 20-50 squares. <10 is small, >60 is commercial/huge.
        if 20 <= squares <= 50:
            size_score = 50
        elif squares > 50:
            size_score = 40 # Good but resource heavy
        else:
            size_score = 20 # Small job

        # 2. Complexity/Waste Score (Waste factor proxy for complexity)
        # Higher waste usually means steeper/complex roof -> Higher margin potential if bid right
        complexity_score = estimates['waste_factor'] * 100 # 0.15 * 100 = 15 points

        heat_score = min(size_score + complexity_score + 20, 100) # +20 base interest
        return int(heat_score)

    def generate_quote(self, project):
        """
        Generates a draft financial quote based on Brain data.
        """
        print(f"CRM: Generating Quote for Project {project.project_id}...")

        if not project.brain_output:
            print("CRM: Missing Brain intelligence. Cannot quote.")
            return None

        estimates = project.brain_output['estimates']
        total_squares = estimates['total_squares_recommended']

        # Financials
        subtotal = total_squares * self.base_price_per_sq
        tax = subtotal * 0.05 # 5% Tax example
        total = subtotal + tax

        heat_score = self.calculate_heat_score(project)

        quote = {
            "project_id": project.project_id,
            "heat_score": heat_score,
            "pipeline_stage": "Quote Ready" if heat_score > 40 else "Nurture",
            "financials": {
                "price_per_sq": self.base_price_per_sq,
                "subtotal": round(subtotal, 2),
                "tax": round(tax, 2),
                "total_estimate": round(total, 2)
            }
        }

        # Update Project State
        project.crm_data = quote
        project.sales_status = quote['pipeline_stage']

        print(f"CRM: Quote Generated. Heat Score: {heat_score}/100. Status: {project.sales_status}")
        return quote

# Initialize the CRM
crm = SalesPipeline()
print("System updated: 'Sales Pipeline' (CRM) logic layer initialized.")
```

```python
# Install necessary libraries for geospatial processing
!pip install rasterio

import os
import json
import requests
import numpy as np
import math
import rasterio
from rasterio.io import MemoryFile

# Configuration Constants
SOLAR_API_BASE = "https://solar.googleapis.com/v1"

# Placeholder for API Key.
# In a real scenario, use: from google.colab import userdata; API_KEY = userdata.get('GOOGLE_API_KEY')
API_KEY = "YOUR_API_KEY_HERE"

class SolarClient:
    """
    Handles interactions with Google Solar API, implementing the 'Weld' phase
    protocols for high-precision data ingestion and verification.
    """
    def __init__(self, api_key):
        self.api_key = api_key

    @staticmethod
    def haversine(lat1, lon1, lat2, lon2):
        """Calculates distance in meters between two lat/lng points."""
        R = 6371000  # Earth radius in meters
        phi1, phi2 = math.radians(lat1), math.radians(lat2)
        dphi = math.radians(lat2 - lat1)
        dlambda = math.radians(lon2 - lon1)
        a = math.sin(dphi/2)**2 + math.cos(phi1) * math.cos(phi2) * math.sin(dlambda/2)**2
        return 2 * R * math.asin(math.sqrt(a))

    def get_building_insights(self, lat, lng):
        """Fetches Building Insights for bounding box verification."""
        url = f"{SOLAR_API_BASE}/buildingInsights:findClosest"
        params = {
            'location.latitude': lat,
            'location.longitude': lng,
            'requiredQuality': 'HIGH',
            'key': self.api_key
        }
        try:
            response = requests.get(url, params=params)
            if response.status_code == 200:
                return response.json()
            else:
                print(f"Insights API Error: {response.status_code} - {response.text}")
                return None
        except Exception as e:
            print(f"Connection Error: {e}")
            return None

    def verify_location(self, insights_data, target_lat, target_lng, max_dist_m=10):
        """
        Verifies if the API returned building is the target by checking centroid distance.
        Returns (True/False, message).
        """
        if not insights_data or 'solarPotential' not in insights_data:
            return False, "No solar data found."

        # Extract bounding box of the whole roof
        bb = insights_data['solarPotential']['wholeRoofStats']['boundingBox']

        # Calculate centroid of the returned bounding box
        center_lat = (bb['lo']['latitude'] + bb['hi']['latitude']) / 2
        center_lng = (bb['lo']['longitude'] + bb['hi']['longitude']) / 2

        dist = self.haversine(target_lat, target_lng, center_lat, center_lng)

        if dist > max_dist_m:
            return False, f"Centroid verification failed: Returned building is {dist:.2f}m away (Limit: {max_dist_m}m)."

        return True, f"Verification Passed: Centroid is {dist:.2f}m away."

    def get_data_layers(self, lat, lng, radius_meters=50):
        """Fetches Data Layers (GeoTIFF URLs) with roofing-specific parameters."""
        url = f"{SOLAR_API_BASE}/dataLayers:get"
        params = {
            'location.latitude': lat,
            'location.longitude': lng,
            'radiusMeters': radius_meters,
            'requiredQuality': 'HIGH',
            'exactQualityRequired': 'true',  # Fail if HIGH not available, avoids merged roofs
            'view': 'FULL_LAYERS',
            'pixelSizeMeters': 0.1,
            'key': self.api_key
        }
        try:
            response = requests.get(url, params=params)
            if response.status_code == 200:
                return response.json()
            elif response.status_code == 404:
                print("Data Layers 404: Location not found or quality requirements not met.")
                return None
            else:
                print(f"Data Layers API Error: {response.status_code} - {response.text}")
                return None
        except Exception as e:
            print(f"Connection Error: {e}")
            return None

    def download_assets(self, layer_urls, output_dir):
        """Downloads GeoTIFFs from the provided URLs to the output directory."""
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)

        downloaded_files = {}
        keys_to_download = ['maskUrl', 'dsmUrl', 'rgbUrl']

        print(f"Downloading assets to {output_dir}...")
        for key in keys_to_download:
            if key in layer_urls:
                url = layer_urls[key]
                try:
                    r = requests.get(url)
                    if r.status_code == 200:
                        filename = f"{key[:-3]}.tiff" # e.g. mask.tiff
                        filepath = os.path.join(output_dir, filename)
                        with open(filepath, 'wb') as f:
                            f.write(r.content)
                        downloaded_files[key] = filepath
                        print(f" - Downloaded {filename}")
                except Exception as e:
                    print(f" - Failed to download {key}: {e}")
        return downloaded_files

class RoofingProject:
    """
    Base class to hold project data as it moves through the pipeline:
    Raw Data -> The Brain -> CRM -> Marketing
    """
    def __init__(self, project_id, lat, lng, api_client=None):
        self.project_id = project_id
        self.location = {"latitude": lat, "longitude": lng}
        self.client = api_client
        self.insights = None
        self.raw_layer_urls = None
        self.local_assets = {}
        self.measurements = {}
        self.sales_status = "New Lead"

    def ingest_data(self):
        """Orchestrates the data ingestion workflow."""
        if not self.client:
            print("No API Client provided.")
            return

        print(f"Starting Ingestion for {self.project_id}...")

        # 1. Get Insights & Verify
        self.insights = self.client.get_building_insights(self.location['latitude'], self.location['longitude'])
        is_verified, message = self.client.verify_location(self.insights, self.location['latitude'], self.location['longitude'])
        print(message)

        if is_verified:
            # 2. Get Data Layers if verified
            self.raw_layer_urls = self.client.get_data_layers(self.location['latitude'], self.location['longitude'])

            if self.raw_layer_urls:
                print("Data Layers URLs acquired.")
                # 3. Download Assets
                asset_dir = f"project_data/{self.project_id}"
                self.local_assets = self.client.download_assets(self.raw_layer_urls, asset_dir)

                # 4. Process/Analyze
                if 'maskUrl' in self.local_assets:
                    self.analyze_roof(self.local_assets['maskUrl'])
            else:
                print("Failed to acquire Data Layers.")
        else:
            print("Skipping Data Layers due to verification failure.")

    def analyze_roof(self, mask_path):
        """Processes the mask GeoTIFF to calculate roof area."""
        try:
            with rasterio.open(mask_path) as src:
                mask = src.read(1) # Read first band (binary)
                # Resolution is usually in the transform or meta (0.1m per pixel if HIGH quality)
                res_x = src.res[0]
                res_y = src.res[1]
                pixel_area = res_x * res_y

                # Count pixels where mask == 1 (Roof)
                roof_pixel_count = np.sum(mask)
                area_m2 = roof_pixel_count * pixel_area

                self.measurements['roof_area_m2'] = area_m2
                print(f"Analysis Complete: Calculated Roof Area: {area_m2:.2f} m²")
        except Exception as e:
            print(f"Error analyzing mask: {e}")

    def __repr__(self):
        return f"Project {self.project_id} @ {self.location} [{self.sales_status}]"

print("System updated: 'Weld' phase complete. Client can now Download & Analyze GeoTIFFs.")
```
